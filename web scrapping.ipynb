# putting all the code together for all item on a single page

```python
all_info = {
    "Items Name": [],
    "New Price" : [],
    "Old Price" : [],
    "Dicounted %": [],
    "Number of people rate" : [],
    "Rated Score": []
}
containers=soup.findAll("div", {"class":"info"})
for container in containers:
    try:
        item_name = container.find("h3", {"class":"name"}).text
        all_info["Items Name"].append(item_name)
    except:
        all_info["Items Name"].append(np.nan)
        
    try:
        item_new_price = container.find("div", class_= "prc").get_text()
        all_info["New Price"].append(item_new_price)
    except:
        all_info["New Price"].append(np.nan)
        
    try:
        item_old_price= container.find("div" ,class_="old").text
        all_info["Old Price"].append(item_old_price)
    except:
        all_info["Old Price"].append(np.nan)
        
    try:
        dicounted_cent = container.find("div", class_="bdg _dsct _sm").text
        all_info["Dicounted %"].append(dicounted_cent)
    except:
        all_info["Dicounted %"].append(np.nan)
        
    try:
        no_of_ppl_rated = container.find("div", class_="rev").text.split("(")[1].strip(")")
        all_info["Number of people rate"].append(no_of_ppl_rated)
    except:
        all_info["Number of people rate"].append(np.nan)
        
    try:
        rated_score = container.find("div", class_="rev").text.split()[0]
        all_info["Rated Score"].append(rated_score)
    except:
        all_info["Rated Score"].append(np.nan)
        
df=pd.DataFrame(all_info)
df
```

# For all the fifty pages

```python

all_info = {
    "Items Name": [],
    "New Price" : [],
    "Old Price" : [],
    "Dicounted %": [],
    "Number of people rate" : [],
    "Rated Score": []
}

page = 1

while page<=5:
    url = f"https://www.jumia.com.ng/catalog/?q=watch&page={page}#catalog-listing"
    html = requests.get(url)
    soup = bs(html.content)

    containers=soup.findAll("div", {"class":"info"})
    for container in containers:
        try:
            item_name = container.find("h3", {"class":"name"}).text
            all_info["Items Name"].append(item_name)
        except:
            all_info["Items Name"].append(np.nan)

        try:
            item_new_price = container.find("div", class_= "prc").get_text()
            all_info["New Price"].append(item_new_price)
        except:
            all_info["New Price"].append(np.nan)

        try:
            item_old_price= container.find("div" ,class_="old").text
            all_info["Old Price"].append(item_old_price)
        except:
            all_info["Old Price"].append(np.nan)

        try:
            dicounted_cent = container.find("div", class_="bdg _dsct _sm").text
            all_info["Dicounted %"].append(dicounted_cent)
        except:
            all_info["Dicounted %"].append(np.nan)

        try:
            no_of_ppl_rated = container.find("div", class_="rev").text.split("(")[1].strip(")")
            all_info["Number of people rate"].append(no_of_ppl_rated)
        except:
            all_info["Number of people rate"].append(np.nan)

        try:
            rated_score = container.find("div", class_="rev").text.split()[0]
            all_info["Rated Score"].append(rated_score)
        except:
            all_info["Rated Score"].append(np.nan)
    page+=1 
df=pd.DataFrame(all_info)
df
```

```python
df.to_csv("jumia_watch.csv", index=False)
```
